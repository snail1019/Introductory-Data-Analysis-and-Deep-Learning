### Phần 4: Trực quan hóa dữ liệu nâng cao


```python

import matplotlib.pyplot as plt
t1_counts = data['T1'].value_counts().sort_index()
plt.plot(t1_counts.index, t1_counts.values)
plt.title('Số lượng học sinh theo từng giá trị T1', fontsize=14)
plt.xlabel('Giá trị T1', fontsize=12)
plt.ylabel('Số lượng học sinh', fontsize=12)
plt.show()
```

```python

# 2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
# a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
# b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
# c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
# d. Từ 8 trở lên = giỏi (ký hiệu “g”)

def classify_t1(x):
    if x < 5:
        return 'y'  # kém
    elif x < 7:
        return 'tb'  # trung bình
    elif x < 8:
        return 'k'  # khá
    else:
        return 'g'  # giỏi

data['phanlopt1'] = data['T1'].apply(classify_t1)

```

```python
# 3. Lập bảng tần số cho biến phanloait1
# Lập bảng tần số cho biến phanlopt1
phanlopt1_freq = data['phanlopt1'].value_counts()
print(phanlopt1_freq)
```

```python
data['phanlopt1'] = pd.cut(
    data['T1'],
    bins=[0, 5, 7, 8, 10],  
    labels=['Y', 'TB', 'K', 'G'],  
    right=False  
)

data[['T1', 'phanlopt1']].head()
phanlopt1_counts = data['phanlopt1'].value_counts()
print(phanlopt1_counts)

t1_grouped = data.groupby(['T1', 'phanlopt1']).size().unstack()
plt.figure(figsize=(8, 5))
t1_grouped.plot(kind='line', linestyle='-', linewidth=2)
plt.legend(title="Phân loại", labels=["Kém (k)", "Trung bình (tb)", "Khá (k)", "Giỏi (g)"])
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
t1_class_counts = data.groupby(['phanlopt1', 'T1']).size().unstack(fill_value=0)
for label in t1_class_counts.index:
    plt.scatter(t1_class_counts.columns, t1_class_counts.loc[label], label=label)  
    for x, y in zip(t1_class_counts.columns, t1_class_counts.loc[label]):
        plt.plot([x, x], [0, y], linestyle='-', color='black', alpha=0.5)  
plt.title('Biểu đồ Drop-line cho T1 theo phân loại phanlopt1', fontsize=14)
plt.xlabel('Giá trị T1', fontsize=12)
plt.ylabel('Số lượng học sinh', fontsize=12)

plt.legend(title='Phân loại T1')
plt.show()
```

```python
# Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối
# 1. Hãy mô tả và khảo sát phân phối cho biến T1
# - Mô tả độ tập trung và phân tán của dữ liệu T1
# - Vẽ biểu đồ Box-Plot và xác định các 10 đại lượng trong biểu đồ đó
# - Mô tả hình dáng lệch của phân phối T1 dựa vào các đại lượng hướng tâm
# - Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
# - Mô tả các đặc trưng của phân phối, mức độ lệch và mức độ nhọn
# - Kiểm chứng phân phối chuẩn QQ-Plot
# - Nhận xét và đánh giá về phân phối của T1
# Mô tả độ tập trung và phân tán của dữ liệu T1
t1_desc = data['T1'].describe()
print(t1_desc)

# Vẽ biểu đồ Box-Plot và xác định các 10 đại lượng trong biểu đồ đó
plt.figure(figsize=(10, 6))
plt.boxplot(data['T1'], vert=False)
plt.title('Box-Plot của T1')
plt.xlabel('Giá trị T1')
plt.show()

# Mô tả hình dáng lệch của phân phối T1 dựa vào các đại lượng hướng tâm
mean_t1 = data['T1'].mean()
median_t1 = data['T1'].median()
mode_t1 = data['T1'].mode()[0]
print(f"Mean: {mean_t1}, Median: {median_t1}, Mode: {mode_t1}")

# Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
plt.figure(figsize=(10, 6))
plt.hist(data['T1'], bins=20, edgecolor='black')
plt.title('Histogram của T1')
plt.xlabel('Giá trị T1')
plt.ylabel('Tần số')
plt.show()

# Mô tả các đặc trưng của phân phối, mức độ lệch và mức độ nhọn
skewness_t1 = data['T1'].skew()
kurtosis_t1 = data['T1'].kurtosis()
print(f"Skewness: {skewness_t1}, Kurtosis: {kurtosis_t1}")

# Kiểm chứng phân phối chuẩn QQ-Plot
import scipy.stats as stats
plt.figure(figsize=(10, 6))
stats.probplot(data['T1'], dist="norm", plot=plt)
plt.title('QQ-Plot của T1')
plt.show()

# Nhận xét và đánh giá về phân phối của T1
# Nhận xét: Dựa vào các biểu đồ và các đại lượng thống kê, ta có thể đánh giá phân phối của T1.
```

```python
import seaborn as sns


# 2. Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
# Gợi ý
# - Trực quan hóa biểu đồ Box-plot, histogram và QQ-plot theo phân nhóm là giá trị
# của ‘phanlopT1’.
import scipy.stats as stats

# Box-plot theo phân nhóm phanlopt1
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopt1', y='T1', data=data)
plt.title('Box-Plot của T1 theo phân nhóm phanlopt1')
plt.xlabel('Phân nhóm T1')
plt.ylabel('Giá trị T1')
plt.show()

# Histogram theo phân nhóm phanlopt1
plt.figure(figsize=(10, 6))
for label in data['phanlopt1'].unique():
    sns.histplot(data[data['phanlopt1'] == label]['T1'], kde=True, label=label, bins=20)
plt.title('Histogram của T1 theo phân nhóm phanlopt1')
plt.xlabel('Giá trị T1')
plt.ylabel('Tần số')
plt.legend(title='Phân nhóm T1')
plt.show()

# QQ-plot theo phân nhóm phanlopt1
plt.figure(figsize=(10, 6))
for label in data['phanlopt1'].unique():
    stats.probplot(data[data['phanlopt1'] == label]['T1'], dist="norm", plot=plt)
    plt.title(f'QQ-Plot của T1 theo phân nhóm {label}')
    plt.show()
```

```python
# 3. Hãy khảo sát tương quan giữa biến DH1 theo biến T1
# Gợi ý
# - Nhận xét giá trị Covariance hoặc Correlation
# - Vẽ biểu đồ Scatter thể hiện liên hệ của biến phụ thuộc DH1 theo biến độc lập T1
# Nhận xét giá trị Covariance hoặc Correlation
correlation = data['DH1'].corr(data['T1'])
print(f"Correlation between DH1 and T1: {correlation}")

# Vẽ biểu đồ Scatter thể hiện liên hệ của biến phụ thuộc DH1 theo biến độc lập T1
plt.figure(figsize=(10, 6))
plt.scatter(data['T1'], data['DH1'], alpha=0.6)
plt.title('Scatter plot of DH1 vs T1')
plt.xlabel('T1')
plt.ylabel('DH1')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 4. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực
# Khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực
for kv in data['KV'].unique():
    subset = data[data['KV'] == kv]
    correlation = subset['DH1'].corr(subset['T1'])
    print(f"Correlation between DH1 and T1 in KV {kv}: {correlation}")

    plt.figure(figsize=(10, 6))
    plt.scatter(subset['T1'], subset['DH1'], alpha=0.6, label=f'KV {kv}')
    plt.title(f'Scatter plot of DH1 vs T1 in KV {kv}')
    plt.xlabel('T1')
    plt.ylabel('DH1')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
# 5. Hãy khảo sát tương quan giữa các biến DH1, DH2, DH3
# Gợi ý
# - Nhận xét ma trận hiệp phương sai hoặc ma trận tương quan
# - Vẽ biểu đồ Scatter giữa các biến
# Nhận xét ma trận hiệp phương sai hoặc ma trận tương quan
correlation_matrix = data[['DH1', 'DH2', 'DH3']].corr()
print("Correlation matrix:")
print(correlation_matrix)

# Vẽ biểu đồ Scatter giữa các biến
pd.plotting.scatter_matrix(data[['DH1', 'DH2', 'DH3']], figsize=(10, 6), alpha=0.6)
plt.suptitle('Scatter matrix of DH1, DH2, and DH3')
plt.show()
```
