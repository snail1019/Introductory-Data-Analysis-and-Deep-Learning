```python
#Câu 9:Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phươngpháp Min-Max Normalization
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv", index_col="STT")
# Tính TBM1, TBM2, TBM3 nếu chưa có
df["TBM1"] = (df["T1"] * 2 + df["L1"] + df["H1"] + df["S1"] + df["V1"] * 2 + df["X1"] + df["D1"] + df["N1"]) / 10
df["TBM2"] = (df["T2"] * 2 + df["L2"] + df["H2"] + df["S2"] + df["V2"] * 2 + df["X2"] + df["D2"] + df["N2"]) / 10
df["TBM3"] = (df["T3"] * 2 + df["L3"] + df["H3"] + df["S3"] + df["V3"] * 2 + df["X3"] + df["D3"] + df["N3"]) / 10

# Chuyển đổi điểm TBM sang thang điểm 4
def convert_to_us_scale(vn_score):
    return (vn_score - 0) / (10 - 0) * (4 - 0) + 0

# Tạo các cột US_TBM1, US_TBM2, US_TBM3
df["US_TBM1"] = convert_to_us_scale(df["TBM1"])
df["US_TBM2"] = convert_to_us_scale(df["TBM2"])
df["US_TBM3"] = convert_to_us_scale(df["TBM3"])

# Kiểm tra kết quả
print(df[["TBM1", "US_TBM1", "TBM2", "US_TBM2", "TBM3", "US_TBM3"]].head())
df.to_csv("dulieuxettuyendaihoc.csv", index=True)
```

```python
#CÂU 10
import pandas as pd

# Đọc dữ liệu
data = pd.read_csv("dulieuxettuyendaihoc.csv", index_col="STT")

# Hàm xác định kết quả xét tuyển
def determine_kqxt(row):
    if row['KT'] in ['A', 'A1']:
        score = (row['DH1'] * 2 + row['DH2'] + row['DH3']) / 4
    elif row['KT'] == 'B':
        score = (row['DH1'] + row['DH2'] * 2 + row['DH3']) / 4
    else:  # Khối khác
        score = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    return 1 if score >= 5.0 else 0

# Áp dụng hàm để tạo cột KQXT
data['KQXT'] = data.apply(determine_kqxt, axis=1)

# In kết quả
print(data.head())
```
